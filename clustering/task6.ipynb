```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Data

```python
dfRtc = pd.read_csv('retail_transactions.csv', encoding='iso-8859-1')
dfRtc.head()
```

```python
dfRtc['item_total'] = dfRtc['UnitPrice'] * dfRtc['Quantity']
dfRtc.head()
```

```python
grouped = dfRtc.groupby('CustomerID').agg({'InvoiceNo':'unique','item_total':'sum'}).reset_index()
dfRtc.rename(columns={'InvoiceNo':'invoice_count','item_total':'amt_total'}, inplace=True)
dfRtc.head()
```

```python
from sklearn.preprocessing import scale
X = scale(dfRtc[['invoice_count','amt_total']])
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('invoice_count')
plt.ylabel('amt_total')
plt.show()
```

```python
from sklearn.cluster import KMeans

#Define
mdlKmcRtc = KMeans(n_clusters=3, random_state=1)

# Fit
mdlKmcRtc.fit(X)

kmeans_lables = mdlKmcRtc.labels_
#Print results
```

```python
#Evauluate using siloette score
from sklearn.metrics import silhouette_score

silhouette_score(X,kmeans_lables)
```

```python
#Function for Kmeans Cluster
def cluster_kMeans(k,x):
    #Define
    mdlKmcRtc = KMeans(n_clusters=k, random_state=1)

    # Fit
    mdlKmcRtc.fit(X)

    kmeans_lables = mdlKmcRtc.labels_
    kmeans_sil_score = silhouette_score(x,kmeans_lables)
    return kmeans_lables, kmeans_sil_score
```

```python
for k in [2,3,4,5,6]:
    kmeans_lable_k , kmeans_sil_score_k = cluster_kMeans(k,X)
    print(f'k: {kmeans_lable_k} \t\t sil_score: {kmeans_sil_score_k}')
```
